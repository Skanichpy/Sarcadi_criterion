# $\underline{Количество\;торговых\;дней}$
### $\cdot$Использование работы с python-фреймами

```python
def DayCount(y1,ticker):
    
    df = pd.read_csv(ticker + '.csv', sep = ';')
    condition = (df["<DATE>"] >= (y1 * 10000)) & (df["<DATE>"]<=((y1+1) * 10000))
    return len(df[condition])

tickers = [
    'WDC',
    'MSFT',
    'INTC',
    'AAPL',
    'ADBE',
    'GLW',
    'GOOG',
    'CSCO',
    'STX',
    'XLNX'
]

years = [iter for iter in range(2012,2020)]

DfDays = pd.DataFrame({
    'Ticker': tickers
})

for year in years:
    
    year_days = []
    
    for ticker in tickers:
        
        year_days.append(DayCount(year,ticker))
        
    DfDays[str(year)] = year_days
    
display(DfDays)
    

```

# $\underline{Максимальное\;относительное\;увеличение\;цены\;акции}$
### $\cdot{В\;следующем\;коде\;данные\;в\;таблице\;представлены\;в\;процентах}$

```python
def MaxRaise(year,ticker):
    df = pd.read_csv(ticker+'.csv', sep = ';')
    condition = (df['<DATE>'] > year * 10000) & (df['<DATE>'] <= (year + 1)*10000)
    TempCloses = list(df[condition]["<CLOSE>"])
    n = len(TempCloses)
    CloseRatio = [round(TempCloses[i+1]/TempCloses[i] - 1, 4)*100 for i in range(0,n-1)]
    return max(CloseRatio)

tickers = [
    'MSFT',
    'INTC',
    'AAPL',
    'ADBE',
    'GLW',
    'GOOG',
    'CSCO',
    'STX'
]

years = [iter for iter in range(2012,2020)]

DfMaxRaise = pd.DataFrame({
    'Ticker':tickers
})



for year in years:
    
    year_maxraise = []
    
    for ticker in tickers:
        
        year_maxraise.append(MaxRaise(year,ticker))
        
    DfMaxRaise[str(year)] = year_maxraise
    
display(DfMaxRaise)
    


```

# $\underline{Логарифмическая\;доходность}$ 

```python
def LogYield(ticker):
    df = pd.read_csv(ticker + '.csv',sep = ';')
    TempClose = list(df['<CLOSE>'])
    LogYield =[0] + [TempClose[i+1]/TempClose[i] for i in range(len(TempClose)-1)]
    return LogYield

```

# $\underline{График\;цены\;закрытия\;акций\;компаний}$

```python
%matplotlib notebook 

t1 = time()
def get_values(ticker, start, stop):
    df = pd.read_csv(ticker + '.csv', sep = ';')
    select = (df['<DATE>'] >= start * 10000) & (df['<DATE>'] < (stop + 1) * 10000)
    return list(df['<CLOSE>'][select])

y1 = 2012
y2 = 2020
comp = input("Введите название компании: ")


y_values = get_values(comp ,y1,y2)
plt.title("График цены закрытия(CLOSE) по годам компании " + str(comp) )
plt.ylabel("Цена закрытия(CLOSE), руб.")
plt.xlabel("Год")
plt.grid()
x_values = [x for x in np.arange(y1,y2-0.0001,(y2-y1)/(len(y_values)))]
plt.xticks([x for x in range(y1,y2)],[x for x in range(y1,y2+1)])
plt.plot(x_values,y_values, color = 'purple')
plt.savefig(comp + '.png')


```
